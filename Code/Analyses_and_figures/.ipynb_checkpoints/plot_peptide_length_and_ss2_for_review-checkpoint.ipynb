```python
record_dict = SeqIO.to_dict(SeqIO.parse('dssp_data/ss.txt',"fasta"))
```

```python
def get_ss2_of_complex(row):
#     print(row)
    if row['pfpd_set']:
        pep_chain = row['peptide']
        rec_chain = row['chain']
        pdb_id = row['pdb_id']
    else:
        pep_chain = row['final_peptide']
        rec_chain = row['final_chain']
        pdb_id = row['final_id']
    pep_query = str(pdb_id.upper()+':'+pep_chain.upper()+':secstr')
    rec_query = str(pdb_id.upper()+':'+rec_chain.upper()+':secstr')
    
    try:
        row['pep_ss2'] = str(record_dict[pep_query].seq)
    except:
        row['pep_ss2'] = 'ss2_missing'
    
    try:
        row['rec_ss2'] = str(record_dict[rec_query].seq)
    except:
        row['rec_ss2'] = 'ss2_missing'
    
    return row
    
```

```python
def replace_8letter_code(ss2):
    ss2 = re.sub('[GI]','H',ss2)
    ss2 = re.sub('[B]','E',ss2)
    ss2 = re.sub('[ST]','C',ss2)
    ss2 = re.sub('[ ]','C',ss2)
    return ss2
```

```python
def define_ss2_groups(ss2):
    if ss2 != 'ss2_missing':

        helix_param = 'H'*helix_length
        strand_param = 'E'*strand_length
        is_helix = helix_param in ss2
        is_strand = strand_param in ss2
       
        if is_helix and is_strand:
            is_combined = True
        else:
            is_combined = False
        
    else:
        is_helix = is_strand = is_combined = None
        
    #print('is helix ', is_helix, ' is_strand ', is_strand, ' is_combined ', is_combined)
    return (is_helix, is_strand, is_combined)
```

```python
ss2_tab = full_tab.apply(lambda row: get_ss2_of_complex(row), axis =1)
ss2_tab['rec_ss2'] = ss2_tab['rec_ss2'].apply(replace_8letter_code)
ss2_tab['pep_ss2'] = ss2_tab['pep_ss2'].apply(replace_8letter_code)
```

```python
pep_ss2_groups = pd.DataFrame(ss2_tab['pep_ss2'].apply(define_ss2_groups).tolist()).rename(columns={0:'pep_is_helix',1:'pep_is_strand',2:'pep_is_combined'})
rec_ss2_groups = pd.DataFrame(ss2_tab['rec_ss2'].apply(define_ss2_groups).tolist()).rename(columns={0:'rec_is_helix',1:'rec_is_strand',2:'rec_is_combined'})
```

```python
tab_with_ss2 = pd.concat([ss2_tab, pep_ss2_groups, rec_ss2_groups], axis=1)
```

```python
summary_tab_with_ss2=tab_with_ss2[['f_id','pdb_id','chain','peptide','final_id','final_chain','final_peptide','pfpd_set','pep_ss2','rec_ss2','rec_is_helix','rec_is_strand','rec_is_combined','pep_is_helix','pep_is_strand','pep_is_combined','peptide_length']]
```

```python
summary_tab_with_ss2['pep_is_coil']=summary_tab_with_ss2.apply(lambda row: not(row['pep_is_helix'] or row['pep_is_strand'] or row['pep_is_combined']), axis=1)
# summary_tab_with_ss2.query('pep_ss2 == "ss2_missing"')['pep_is_coil'] = None
```

```python
summary_tab_with_ss2[summary_tab_with_ss2.pep_ss2 == 'ss2_missing'] = "None"
```

```python
summary_tab_with_ss2.to_csv('improved_data_set_26092021_with_pep_length_and_ss2_from_dssp.tsv',sep='\t',index=False)
```
