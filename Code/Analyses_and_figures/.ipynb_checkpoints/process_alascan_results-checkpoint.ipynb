```python
def return_best_combined(row):
    if row['rmsBB_if_x'] <= row['rmsBB_if_y']:
        row['combined_rmsBB_if'] = row['rmsBB_if_x']
        row['best_run_type'] = row['run_type_x']
    else:
        row['combined_rmsBB_if'] = row['rmsBB_if_y']
        row['best_run_type'] = row['run_type_y']
    return(row)
    
```

```python
combined_rms_short = combined_rms_short.apply(lambda row: return_best_combined(row), axis=1)
# combined_rms_short['combined_rmsBB_if'] = combined_rms_short.apply(lambda row: min(row['rmsBB_if_x'], row['rmsBB_if_y']), axis=1)
combined_rms_short.rename(columns={'pdb_id':'complex'}, inplace= True)
combined_rms_short.replace('approved', 'Approved', inplace=True)
```

```python
def process_alascan_tabs(complex_i, run_type):
#     print(complex_i)
    native_path = glob.glob(str('*'+run_type+'*native*'+complex_i+'*'))[0]
    model_path = glob.glob(str('*'+run_type+'*best*'+complex_i+'*'))[0]

    merged_tab = merge_tables(native_path, model_path)
#     merged_tab['run_type'] = run_type.strip('*')
    merged_tab['run_type'] = run_type
    merged_tab['complex'] = complex_i
    
    return merged_tab
    
    
```

```python
def merge_tables(native_path, model_path):
    native_tab = pd.read_table(native_path, sep='\s+', skiprows=2)
    model_tab = pd.read_table(model_path, sep='\s+', skiprows=2)
#     print(len(native_tab))
#     print(len(model_tab))
    native_tab.drop(['pdb#','int_id','DDG(complex,obs)'],axis=1, inplace=True)
    model_tab.drop(['pdb#','int_id','DDG(complex,obs)'],axis=1, inplace=True)
    
    merged_tab = native_tab.merge(model_tab, how='outer', on=['res#','chain','aa'])
    merged_tab.rename(columns={'DDG(complex)_x':'DDG_complex_native','DDG(complex)_y':'DDG_complex_model'}, inplace=True)
    merged_tab.rename(columns={'DG(partner)_x':'DG_partner_native','DG(partner)_y':'DG_partner_model'}, inplace=True)

#     print(len(merged_tab))

    return(merged_tab)
```

```python
def floor_rms_values(val):
    val_bins = [2.5, 5]
    if val <= val_bins[0]:
        i_bin = "x <= 2.5"
#         i_bin = val_bins[0]
    elif val <= val_bins[1]:
        i_bin = "2.5 < x <= 5"
#         i_bin = val_bins[1]
    else:
        i_bin = "x > 5"
#         i_bin = 20
        
    return i_bin
```

```python
def kendall_pval(x,y):
    return kendalltau(x,y)[1]
    
def pearsonr_pval(x,y):
    return pearsonr(x,y)[1]
    
def spearmanr_pval(x,y):
    return spearmanr(x,y)[1]
    
```

```python
def compute_chi_values(contingency_mat):
    chi_vals = chi2_contingency(contingency_mat)
    return chi_vals
    
```

```python
def compute_accuracy_metrics(i_tab, filt_dg_partner, hotspot_threshold, rms_threshold):
    
    if filt_dg_partner == 0:
        i_tab = i_tab
    elif filt_dg_partner == 1:
        i_tab = i_tab.query('DG_partner_native < 1').query('DG_partner_model < 1')
    
    
    i_tab = i_tab.query("combined_rmsBB_if <= @rms_threshold")
    chain_list = ["A","B"]
    partner_list = ["receptor","peptide"]

    set_list = ["Motif","Non-motif",'Approved']
    bin_list = ['x <= 2.5', '2.5 < x <= 5', 'x > 5']
    
    value_list = []
    tmp_value_list = []
    
    for i_chain in enumerate(chain_list):
        for k_set in enumerate(set_list):
            tmp_value_list = []
            filt_tab = i_tab.query('chain == @i_chain[1]').query('dataset == @k_set[1]')
            tmp_value_list.append(len(filt_tab))
            tmp_value_list.append(filt_tab[['DDG_complex_native', 'DDG_complex_model']].corr(method = 'pearson').iloc[1][0])
            tmp_value_list.append(filt_tab[['DDG_complex_native', 'DDG_complex_model']].corr(method = pearsonr_pval).iloc[1][0])
            
            tmp_value_list.append(filt_tab[['DDG_complex_native', 'DDG_complex_model']].corr(method = 'spearman').iloc[1][0])
            tmp_value_list.append(filt_tab[['DDG_complex_native', 'DDG_complex_model']].corr(method = spearmanr_pval).iloc[1][0])
            
            filt_tab['native_label'] = filt_tab['DDG_complex_native'].apply(lambda x: 1 if x>=hotspot_threshold else 0)
            filt_tab['model_label'] = filt_tab['DDG_complex_model'].apply(lambda x: 1 if x>=hotspot_threshold else 0)
#             print(filt_tab)
            confusion_mat = confusion_matrix(filt_tab['native_label'], filt_tab['model_label'])
#             tmp_value_list.append(confusion_mat)
            tn, fp, fn, tp = confusion_mat.ravel()
            contig_mat = np.array([[fn, fp], [tn, tp]])
            tmp_value_list.append(contig_mat)
            tmp_value_list.extend([tn, fp, fn, tp])
            chi_results = compute_chi_values(contig_mat)
            chi_statistic = chi_results[0]
            chi_pval = chi_results[1]
            chi_expected = chi_results[3]
            tmp_value_list.append(chi_statistic)
            tmp_value_list.append(chi_pval)
            tmp_value_list.append(chi_expected)
            tmp_value_list.append(k_set[1])
            tmp_value_list.append(partner_list[i_chain[0]])
            tmp_value_list.append(i_chain[1])

            value_list.append(tmp_value_list)
#     axs[i_chain[0]][k_set[0]].set_title(str(k_set[1])+', '+str(partner_list[i_chain[0]]))

    colnames_list = ['total_size','pearson_r','pearson_r_pval','spearman_r','spearman_r_pval','contingency_mat','TN','FP','FN','TP','chi_statistic','chi_pval','chi_expected','dataset','partner', 'partner_chain']
    df = pd.DataFrame(value_list)
    df.columns = colnames_list
    df['rms_threshold'] = rms_threshold
    df['hotspot_threshold'] = hotspot_threshold
    
    return df
    
    
```

```python
rms_threshold_list = [2.5, 5, 1000]
accuracy_df_list = [compute_accuracy_metrics(merged_with_rmsd, 0, 1.5, x) for x in rms_threshold_list]
accuracy_df_list.extend([compute_accuracy_metrics(merged_with_rmsd, 0, 1, x) for x in rms_threshold_list])


concated_dfs = pd.concat(accuracy_df_list)
# concated_dfs.to_csv('concated_accuracy_dfs.tsv',sep='\t',index=False)

```
