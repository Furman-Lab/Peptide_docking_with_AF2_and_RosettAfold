```python
import pandas as pd 
```

```python
struct_based = pd.read_table("ziv_structural_fusion_mapping_curated.tsv",sep="\t")
sifts_based = pd.read_table("curated_multi_uniprot_matched_ecod_families_by_sifts_for_fusion_mapping.tsv",sep="\t")
```

```python
struct_based
```

```python
short_struct_based=struct_based[['f_id','rep_pdb_id','rep_chain','rep_peptide_id','hom_pdb_id','hom_chain_id','single_chain_in_peptide_location?']]
short_struct_based.rename(columns={'single_chain_in_peptide_location?':'single_chain_in_peptide_location'}, inplace=True)
```

```python
short_struct_based=short_struct_based.query('single_chain_in_peptide_location != "no"')
short_struct_based=short_struct_based.query('single_chain_in_peptide_location != "prob_no"')
short_struct_based.dropna(subset=['single_chain_in_peptide_location'], inplace=True)
short_struct_based.sort_values(by=['single_chain_in_peptide_location'], ascending=False, inplace=True)
short_struct_based.drop_duplicates(subset=['rep_pdb_id'], inplace=True)
```

```python
short_sifts = sifts_based[['pdb','chain','f_id','final_id','final_chain','final_peptide','single_chain_in_peptide_location?']]
short_sifts.rename(columns={'single_chain_in_peptide_location?':'single_chain_in_peptide_location'}, inplace=True)
```

```python
short_sifts=short_sifts.query('single_chain_in_peptide_location != "no"')
short_sifts=short_sifts.query('single_chain_in_peptide_location != "prob_not"')
short_sifts=short_sifts.query('single_chain_in_peptide_location != "prob_no"')
short_sifts=short_sifts.query('single_chain_in_peptide_location != "no "')
short_sifts.dropna(subset=['single_chain_in_peptide_location'], inplace=True)
short_sifts.sort_values(by=['single_chain_in_peptide_location'], ascending=False, inplace=True)
short_sifts.drop_duplicates(subset=['final_id'], inplace=True)
```

```python
short_struct_based.rename(columns={'rep_pdb_id':'final_id','hom_pdb_id':'homolog_pdb'},inplace=True)
short_sifts.rename(columns={'pdb':'homolog_pdb'},inplace=True)
```

```python
memorization_list = short_sifts.final_id.to_list()
# memorization_list.extend(short_struct_based.rep_pdb_id.to_list())
memorization_list.extend(short_struct_based.final_id.to_list())

uniq_memorization_list = set(memorization_list)
```

```python
# Use to write the list of 
pd.Series(list(uniq_memorization_list)).to_csv('memorization_list_of_pdbs.tsv',sep='\t', index=False)
```

```python
merged_memorization_tab = short_sifts[['f_id','final_id','homolog_pdb']].merge(short_struct_based[['f_id','final_id','homolog_pdb']], how='outer')
merged_memorization_tab.to_csv('memorization_tab_including_pdbs_and_homologs.tsv',sep='\t', index=False)
```
